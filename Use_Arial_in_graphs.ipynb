```python
# Arialフォントファイルをダウンロード
!wget -O Arial.ttf "https://github.com/matomo-org/travis-scripts/raw/master/fonts/Arial.ttf"

# 必要なライブラリをインポート
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# フォントファイルをmatplotlibのフォントマネージャに追加
fm.fontManager.addfont('./Arial.ttf')

# Arialをデフォルトフォントに設定
plt.rcParams['font.family'] = 'Arial'

# テスト用のグラフを作成
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Arial Font Test - Sample Graph')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()

# フォントが正しく設定されているか確認
print(f"Current font family: {plt.rcParams['font.family']}")
```

```python
# Times New Romanフォントファイルをダウンロード
!wget -O TimesNewRoman.ttf "https://github.com/justrajdeep/fonts/raw/master/Times New Roman.ttf"

# 必要なライブラリをインポート
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

# フォントファイルをmatplotlibのフォントマネージャに追加
fm.fontManager.addfont('./TimesNewRoman.ttf')

# Times New Romanをデフォルトフォントに設定
plt.rcParams['font.serif'] = "Times New Roman"
plt.rcParams['font.family'] = "serif"

# テスト用のグラフを作成
import numpy as np
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y)
plt.title('Times New Roman Font Test - Sample Graph')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()

# フォントが正しく設定されているか確認
print(f"Current font family: {plt.rcParams['font.family']}")
print(f"Current serif font: {plt.rcParams['font.serif']}")
```
