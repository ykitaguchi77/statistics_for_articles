```python
# CSVの形が特殊なので1行ずつ開く。


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv

def extract_refractive_data_to_df(file_path):
    # セクションの開始行と終了行を探す
    section_start = None
    data_start = None
    section_end = None

    with open(file_path, 'r', encoding='utf-8', errors='replace') as file:
        lines = file.readlines()

        for i, line in enumerate(lines):
            if '[Refractive Keratometric]' in line:
                section_start = i
            elif section_start is not None and 'Data' in line and data_start is None:
                data_start = i + 1  # Dataの次の行からデータ開始
            elif section_start is not None and data_start is not None and line.strip() and line.strip()[0] == '[':
                # 次のセクションの開始
                section_end = i
                break

    if section_start is None or data_start is None:
        print("必要なセクションが見つかりませんでした。")
        return None

    if section_end is None:
        section_end = len(lines)  # ファイルの終わりまで

    # データ行を抽出して処理
    data_rows = []
    for i in range(data_start, section_end):
        line = lines[i].strip()
        if not line:
            continue

        # CSVの行を処理
        row_values = []
        for value in line.split(','):
            value = value.strip()
            if value:
                try:
                    row_values.append(float(value))
                except ValueError:
                    # 数値に変換できない場合はスキップ
                    continue

        if row_values:
            data_rows.append(row_values)

    # すべての行が同じ長さになるように調整
    max_length = max(len(row) for row in data_rows) if data_rows else 0

    for i in range(len(data_rows)):
        if len(data_rows[i]) < max_length:
            # 不足している要素をNaNで埋める
            data_rows[i] = data_rows[i] + [np.nan] * (max_length - len(data_rows[i]))

    # DataFrameに変換
    df = pd.DataFrame(data_rows)

    return df

# ファイルパス
csv_dir = "/content/drive/Shareddrives/岩崎Dr_IgG4 deulk/眼瞼下垂/CASIAデータ/上皮/01932018_20191211_111300_CASIA2_R_002_MAP(Data).csv"

# データを抽出してDataFrameに変換
df = extract_refractive_data_to_df(csv_dir)
df
```

##**左右による補正**

```python
import pandas as pd

# ------------------ 方向ごとの j インデックス ------------------
J_IDX = {
    "R": {"N":0, "SN":1, "S":2, "ST":3, "T":4, "IT":5, "I":6, "IN":7},
    "L": {"N":4, "SN":3, "S":2, "ST":1, "T":0, "IT":7, "I":6, "IN":5},
}

ORDER = ["N", "SN", "S", "ST", "T", "IT", "I", "IN"]        # 常にこの順

# ------------------ サマリー作成関数 ------------------
def build_zone_summary(raw_df: pd.DataFrame, side: str = "R") -> pd.DataFrame:
    """
    raw_df : Sec(i,j) を列にもつ 1 行 DataFrame（あなたの result_df）
    side   : "R" または "L"
    return : 指定フォーマットの 1 行 DataFrame
    """
    side = side.upper()
    if side not in ("R", "L"):
        raise ValueError("side must be 'R' or 'L'")

    # 出力カラム
    cols_out = (
        ['Sec(0-2mm)'] +
        [f'Sec(2-5mm)<{d}>'  for d in ORDER] +
        [f'Sec(5-7mm)<{d}>'  for d in ORDER] +
        [f'Sec(7-10mm)<{d}>' for d in ORDER]
    )
    out = pd.DataFrame(columns=cols_out, index=[0])

    # ------ 中心 ------
    out.loc[0, 'Sec(0-2mm)'] = raw_df.loc[0, 'Sec(0,0)']

    # ------ 各リング ------
    for ring, label in zip([1, 2, 3],
                           ['Sec(2-5mm)', 'Sec(5-7mm)', 'Sec(7-10mm)']):
        for d in ORDER:
            j = J_IDX[side][d]
            out.loc[0, f'{label}<{d}>'] = raw_df.loc[0, f'Sec({ring},{j})']

    return out
```

```python
summary_R = build_zone_summary(result_df, side="R")  # 右眼
summary_L = build_zone_summary(result_df, side="L")  # 左眼
print(summary_R)
```

#**複数ファイル一括処理**

```python
import os
import pandas as pd
import numpy as np
from tqdm import tqdm

# --- CSV読込関数 ---
def extract_refractive_data_to_df(file_path):
    section_start = None
    data_start = None
    section_end = None

    with open(file_path, 'r', encoding='utf-8', errors='replace') as file:
        lines = file.readlines()

        for i, line in enumerate(lines):
            if '[Refractive Keratometric]' in line:
                section_start = i
            elif section_start is not None and 'Data' in line and data_start is None:
                data_start = i + 1
            elif section_start is not None and data_start is not None and line.strip() and line.strip()[0] == '[':
                section_end = i
                break

    if section_start is None or data_start is None:
        return None

    if section_end is None:
        section_end = len(lines)

    data_rows = []
    for i in range(data_start, section_end):
        line = lines[i].strip()
        if not line:
            continue
        row_values = []
        for value in line.split(','):
            value = value.strip()
            if value:
                try:
                    row_values.append(float(value))
                except ValueError:
                    continue
        if row_values:
            data_rows.append(row_values)

    max_length = max(len(row) for row in data_rows) if data_rows else 0
    for i in range(len(data_rows)):
        if len(data_rows[i]) < max_length:
            data_rows[i] += [np.nan] * (max_length - len(data_rows[i]))

    return pd.DataFrame(data_rows)

# --- セクション定義 ---
sections = {
    'Sec(0,0)': {'rows': [(0, 32)],                'start_col':  0, 'num_cols': 49},
    'Sec(1,0)': {'rows': [(0, 2), (30, 32)],       'start_col': 49, 'num_cols': 75},
    'Sec(1,1)': {'rows': [(2, 6)],                 'start_col': 49, 'num_cols': 75},
    'Sec(1,2)': {'rows': [(6, 10)],                'start_col': 49, 'num_cols': 75},
    'Sec(1,3)': {'rows': [(10, 14)],               'start_col': 49, 'num_cols': 75},
    'Sec(1,4)': {'rows': [(14, 18)],               'start_col': 49, 'num_cols': 75},
    'Sec(1,5)': {'rows': [(18, 22)],               'start_col': 49, 'num_cols': 75},
    'Sec(1,6)': {'rows': [(22, 26)],               'start_col': 49, 'num_cols': 75},
    'Sec(1,7)': {'rows': [(26, 30)],               'start_col': 49, 'num_cols': 75},
    'Sec(2,0)': {'rows': [(0, 2), (30, 32)],       'start_col':124, 'num_cols': 50},
    'Sec(2,1)': {'rows': [(2, 6)],                 'start_col':124, 'num_cols': 50},
    'Sec(2,2)': {'rows': [(6, 10)],                'start_col':124, 'num_cols': 50},
    'Sec(2,3)': {'rows': [(10, 14)],               'start_col':124, 'num_cols': 50},
    'Sec(2,4)': {'rows': [(14, 18)],               'start_col':124, 'num_cols': 50},
    'Sec(2,5)': {'rows': [(18, 22)],               'start_col':124, 'num_cols': 50},
    'Sec(2,6)': {'rows': [(22, 26)],               'start_col':124, 'num_cols': 50},
    'Sec(2,7)': {'rows': [(26, 30)],               'start_col':124, 'num_cols': 50},
    'Sec(3,0)': {'rows': [(0, 2), (30, 32)],       'start_col':174, 'num_cols': 75},
    'Sec(3,1)': {'rows': [(2, 6)],                 'start_col':174, 'num_cols': 75},
    'Sec(3,2)': {'rows': [(6, 10)],                'start_col':174, 'num_cols': 75},
    'Sec(3,3)': {'rows': [(10, 14)],               'start_col':174, 'num_cols': 75},
    'Sec(3,4)': {'rows': [(14, 18)],               'start_col':174, 'num_cols': 75},
    'Sec(3,5)': {'rows': [(18, 22)],               'start_col':174, 'num_cols': 75},
    'Sec(3,6)': {'rows': [(22, 26)],               'start_col':174, 'num_cols': 75},
    'Sec(3,7)': {'rows': [(26, 30)],               'start_col':174, 'num_cols': 75},
}

# --- 指定方向のインデックス ---
J_IDX = {
    "R": {"N":0, "SN":1, "S":2, "ST":3, "T":4, "IT":5, "I":6, "IN":7},
    "L": {"N":4, "SN":3, "S":2, "ST":1, "T":0, "IT":7, "I":6, "IN":5},
}
ORDER = ["N", "SN", "S", "ST", "T", "IT", "I", "IN"]

# --- ゾーン集約関数 ---
def build_zone_summary(raw_df: pd.DataFrame, side: str) -> pd.DataFrame:
    cols_out = (
        ['filename', 'side', 'Sec(0-2mm)'] +
        [f'Sec(2-5mm)<{d}>' for d in ORDER] +
        [f'Sec(5-7mm)<{d}>' for d in ORDER] +
        [f'Sec(7-10mm)<{d}>' for d in ORDER]
    )
    out = pd.DataFrame(columns=cols_out, index=[0])
    out['side'] = side

    out['Sec(0-2mm)'] = raw_df.loc[0, 'Sec(0,0)']
    for ring, label in zip([1, 2, 3], ['Sec(2-5mm)', 'Sec(5-7mm)', 'Sec(7-10mm)']):
        for d in ORDER:
            j = J_IDX[side][d]
            out[f'{label}<{d}>'] = raw_df.loc[0, f'Sec({ring},{j})']
    return out


def process_all_csvs_with_options(parent_dir, limit=None, output_path=None):
    summary_rows = []
    file_list = []

    # ファイル一覧の取得とフィルタリング
    for root, _, files in os.walk(parent_dir):
        for fname in files:
            if fname.lower().endswith(".csv") and "CASIA2_" in fname:
                file_list.append(os.path.join(root, fname))
    if limit is not None:
        file_list = file_list[:limit]

    # tqdmで進捗表示
    for fpath in tqdm(file_list, desc="Processing CSV files"):
        fname = os.path.basename(fpath)
        side_marker = fname.split("CASIA2_")[1][0]
        side = "R" if side_marker == "R" else "L"

        df_raw = extract_refractive_data_to_df(fpath)
        if df_raw is None:
            continue

        # 平均値計算
        result_df = pd.DataFrame(columns=[f'Sec({i},{j})' for i in range(4) for j in range(8)])
        result_df.loc[0] = np.nan
        for col_name, s in sections.items():
            start_c, end_c = s['start_col'], s['start_col'] + s['num_cols']
            vals = []
            for r0, r1 in s['rows']:
                sub = df_raw.iloc[r0:r1, start_c:end_c]
                vals.extend(pd.to_numeric(sub.stack(), errors='coerce').dropna())
            if vals:
                result_df.loc[0, col_name] = np.mean(vals)

        summary = build_zone_summary(result_df, side)
        summary['filename'] = fname
        summary_rows.append(summary)

    # 結合と保存
    final_df = pd.concat(summary_rows, ignore_index=True) if summary_rows else pd.DataFrame()
    if output_path:
        final_df.to_csv(output_path, index=False)

    return final_df

```
